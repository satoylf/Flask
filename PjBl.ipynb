## Projeto:
- Definição da ideia de aplicação do grupo para a disciplina.

 - Automação de processos na Agricultura
 - Nossa fazenda é automaticamente monitorada por sensores de temperatura, pressão e umidade, com uma interface web disponível para operadores e administradores monitorarem de forma rápida e prática o estado atual das plantações, com um sistema para alertar possíveis condições climáticas que possam prejudicar o cultivo da plantação.
- Quais os dados serão necessários para o desenvolvimento da aplicação.

 - Registro do tempo (temperatura/umidade/pressão), Condições pra cada planta,
   Usuários (permissões/senha/...)

## Framework de Design:
- Definir qual framework de design será utilizado no projeto.

 - [Bootstrap ](https://getbootstrap.com/docs/5.2/getting-started/javascript/)

- Apresentar as suas características principais.

 - Customização rápida e com alta responsividade, melhorando o desenvolvimento da aplicação

- Elencar a razão da escolha (alguém do grupo já utilizou, fácil de trabalhar, opções de templates mais interessantes etc.).
 
 - Foi escolhido por parecer o mais prático para se trabalhar, mesmo que ninguem do grupo seja familiarizado com o framework
